```python
count = df['ZipCode'].value_counts() 
print(count)
```

```python
# Replacing categorial data to integer. lengthy process- use label encoder
mapping = {'a': 0, 'b': 1, 'p': 0, 'g': 1, 't': 0, 'f': 1, 's': 1, 'u': 0, 'y': 1}
df.replace({'Male': mapping, 'DriversLicense': mapping, 'BankCustomer': mapping, 'PriorDefault': mapping, 'Employed': mapping, 'Citizen': mapping, 'Married': mapping})
```

```python
df = df.replace('?',np.NaN)
```

```python
# This will replace the NaN with the above column value
df = df.fillna(method='ffill')
print(df.count())
df.tail(20)
```

```python
df.columns
```

```python
#Import Label encoder to convert object 
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Male']= label_encoder.fit_transform(df['Male'])

label_encoder = preprocessing.LabelEncoder()
df['Married']= label_encoder.fit_transform(df['Married'])

label_encoder = preprocessing.LabelEncoder()
df['EducationLevel']= label_encoder.fit_transform(df['EducationLevel'])

label_encoder = preprocessing.LabelEncoder()
df['Ethnicity']= label_encoder.fit_transform(df['Ethnicity'])

label_encoder = preprocessing.LabelEncoder()
df['PriorDefault']= label_encoder.fit_transform(df['PriorDefault'])

label_encoder = preprocessing.LabelEncoder()
df['Employed']= label_encoder.fit_transform(df['Employed'])

label_encoder = preprocessing.LabelEncoder()
df['DriversLicense']= label_encoder.fit_transform(df['DriversLicense'])

label_encoder = preprocessing.LabelEncoder()
df['Citizen']= label_encoder.fit_transform(df['Citizen'])

label_encoder = preprocessing.LabelEncoder()
df['BankCustomer']= label_encoder.fit_transform(df['BankCustomer'])

label_encoder = preprocessing.LabelEncoder()
df['Approved']= label_encoder.fit_transform(df['Approved'])
```

```python
print(df.count())
df.head(20)
```

```python
# Drop coulmns from a data frame in pandas
df1 = df.drop(['ZipCode', 'DriversLicense',], axis=1)
```

```python
# Creating new vRIABLE FOR input features and labels

#X = df["Male","Age","Debt","Married","BankCustomer","EducationLevel","Ethnicity","YearsEmployed","PriorDefault","Employed","CreditScore","Citizen","Income"]
from sklearn.model_selection import train_test_split
y = df1["Approved"]
X = df1.drop(['Approved'], axis=1)
# Splitting the data into training and testing set

X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size = 0.2,
                                                   random_state = 123)
X.shape
```

```python
# Import MinMax Scaler
from sklearn.preprocessing import MinMaxScaler

# Scaling X_train and X_test
scaler =  MinMaxScaler(feature_range=(0, 1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.transform(X_test)
```

```python
rescaledX = scaler.transform(X)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
rf = RandomForestClassifier(n_estimators=500)
rf.fit(rescaledX_train,y_train)
y_pred = rf.predict(rescaledX_test)
print("Random Forest Classifier has accuracy of: ",rf.score(rescaledX_test,y_test))
confusion_matrix(y_test,y_pred)
```

```python
# Import logisticRegression
from sklearn.linear_model import LogisticRegression

# Fitting logistic regression with default parameter values
LogisticRegression = LogisticRegression()
LogisticRegression.fit(rescaledX_train,y_train)
```

```python
LogisticRegression.score(rescaledX_test,y_test)
```

```python
y_pred = LogisticRegression.predict(rescaledX_test)
y_pred
```

```python
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1
```

```python
# Import Confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

#Using the trained model to predict instances from the test set
y_pred = LogisticRegression.predict(rescaledX_test)

# Getting the accuracy score of predictive model
print("Logistic Regression classifier has accuracy of: ", LogisticRegression.score(rescaledX_test,y_test))

#Evaluate the confusion Matrix
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
